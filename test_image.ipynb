```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import glob
from natsort import natsorted
```

```python
# get file from folder
image_paths = natsorted(glob.glob(".\datasets\Contraste de Fase\*"))
# load the images and convert them to to grayscale
# save image to new folder but keep the name of the image
for image_path in image_paths:
    image = Image.open(image_path)
    image = np.array(image)[500:, 500:950, 0]
    image = Image.fromarray(image)
    image.save("./datasets/Contraste de Fase Gray/" + image_path.split("\\")[-1])
```
